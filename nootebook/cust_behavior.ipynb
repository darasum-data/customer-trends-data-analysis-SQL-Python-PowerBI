## Missing Data Handling: 
Checked for null values and imputed missing value in the Review Rating column using the median rating of each product category.

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

## Column Standardization: 
Renamed column to snake case  for better readability and documents.

```python

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df.head()
```

## â€¢	Feature Engineering:

Create age_group column by binning customer ages.

```python
# Create a column age_group feature engineering
age_labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_labels)

df[['age', 'age_group']].head(10)
```

Create purchase_frequency_days columns from purchase data.

```python
# Create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly' : 7,
    "Monthly" : 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months" : 90
}

df['purchase_frequency_days'] = df["frequency_of_purchases"].map(frequency_mapping)

# df[df['purchase_frequency_days'].isnull()]
df['purchase_frequency_days'].isnull().value_counts()
```

## Data Consistency Check: 
Verified if discount_applied and promo_code_used were redundant; drop promo_code_used.

```python
# check redunency columns
df[['discount_applied', 'promo_code_used']].head(10)

# (df['promo_code_used'] == df['discount_applied']).all()
df['promo_code_used'].equals (df['discount_applied'])
```

```python
df = df.drop('promo_code_used', axis=1)
df.columns
```
